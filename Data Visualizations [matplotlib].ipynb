# Matplotlib

```python
from matplotlib import pyplot as plt
from collections import Counter
```

### Plot charts

```python
years =[1950,1960,1970,1980,1990,2000,2010,2020,2021]
gdp=[300.2,543.3,1075.9,2862.5,5979.6,10289.7,14958.3,16502.5,17586.9]
```

```python
#for making color use color =' ' or 'g','b' and 'r' ...etc with 'g-' or 'g:' means solid, dashed, dotted
graph = plt.plot(years,gdp,color='red',marker='o',linestyle='dashed')
graph= plt.title("years of residence")
plt.ylabel("years of $")
graph= plt.show()
```

### Bar charts

```python
Movies =["Annie mall","Ben-Hur","Casablanca","Candhl","West sides story"]
num_oscars=[5,11,3,8,18]
plt.bar(range(len(Movies)),num_oscars,color='yellow')
plt.title("My favourite movies")
plt.ylabel("Academy awards")
```

### Histogram

```python
grades =[83,95,91,87,78,8,85,82,100,67,73,77,0]

histogram = Counter(min(grade // 10 * 10,90)for grade in grades)
plt.bar([x + 5 for x in histogram.keys()],histogram.values(),10,edgecolor=(0,0,0),color='orange')
plt.axis([-5,105,0,5])
plt.xticks([10 * i for i in range(11)])
plt.xlabel("Decile")
plt.ylabel("num of students")
plt.title("Distribution of Exam 1 Grades")
plt.show()
```

## Line charts

```python
variance = [1,2,4,8,16,32,64,128,256]
bias_squared = [256,128,64,32,16,8,4,2,1]
total_error =[ x + y for x, y in zip(variance,bias_squared)]
xs = [i for i,_ in enumerate(variance)]

#we can make multiple calls to plt.plto
#to show multiple series on the same chart
plt.plot(xs,variance, 'g-',label='variance') #green solid line
plt.plot(xs,blas_squared, 'r-',label='blas^2') #red, dot-dashed line
plt.plot(xs,total_error,'b:',label='total error') #blue dotted line

plt.legend(loc=9)
plt.xlabel("Model complexity")
plt.xticks([])
```

### Scatterplots

```python
friends = [70,65,72,63,71,64,69,64,67]
minutes = [175,170,205,120,220,130,105,145,190]
labels =  ['a','b','c','d','e','f','g','h','i']
plt.scatter(friends,minutes)

#label each point
for label, friend_count, minute_count in zip(labels,friends,minutes):
    plt.annotate(label,xy=(friend_count,minute_count), #put the lable with its points
                xytext=(5,-5), #put slightly offset
                textcoords='offset points')
    
plt.title("Dally Minutes VS, number of friends")
plt.xlabel("of friends")
plt.ylabel("daily minutes spent on sites")
plt.show()
```
